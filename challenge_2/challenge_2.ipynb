```python
import pandas as pd
from collections import Counter
```

```python
order = pd.read_csv('association_order.csv')
rules = pd.read_csv('rules.csv')
```

```python
def rule_func(x):
    # 3 item rule
    if '&' in x:
        rule = x.split('>')
        if '&' in rule[0]:
            rule_left = rule[0].split('&')
            rule_right = rule[1]

            items_1 = order.loc[(order['itemid'] == int(rule_left[0])) | (order['itemid'] == int(rule_left[1]))]
            items_2 = order.loc[(order['itemid'] == int(rule_left[0])) | (order['itemid'] == int(rule_left[1])) | (order['itemid'] == int(rule_right))]

            items_1 = [item for item, count in Counter(items_1['orderid']).items() if count > 1]
            items_2 = [item for item, count in Counter(items_2['orderid']).items() if count > 2]

            try:
                return int( ( len(items_2) / len(items_1) ) * 1000 )
            except:
                return 0

            
        else:
            rule_left = rule[0]
            rule_right = rule[1].split('&')

            items_1 = order.loc[order['itemid'] == int(rule_left)]
            items_2 = order.loc[(order['itemid'] == int(rule_right[0])) | (order['itemid'] == int(rule_right[1])) | (order['itemid'] == int(rule_left))
]
            items_2 = [item for item, count in Counter(items_2['orderid']).items() if count > 2]

            try:
                return int( ( len(items_2) / len(items_1) ) * 1000 )
            except:
                return 0
    # 2 item rule
    else:
        rule = x.split('>')

        items_1 = order.loc[order['itemid'] == int(rule[0])]
        items_2 = order.loc[(order['itemid'] == int(rule[0])) | (order['itemid'] == int(rule[1]))]
        

        items_2 = [item for item, count in Counter(items_2['orderid']).items() if count > 1]
        
        try:
            return int( ( len(items_2) / len(items_1) ) * 1000 )
        except:
            return 0
```

```python
rules['confidence'] = rules['rule'].apply(rule_func)
```
